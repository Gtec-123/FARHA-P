DATA PREPROCESSING

ONE-HOT-ENCODER


```python
# Drop identifier columns
x_preprocessed = x.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

# One-hot encode categorical features
x_preprocessed = pd.get_dummies(x_preprocessed, columns=['Geography', 'Gender'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_preprocessed, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_scaled = sc.fit_transform(x_train)
# Select numerical columns for scaling
numerical_cols = ['CreditScore', 'Age', 'Tenure']
# Fit and transform the scaler on the training data
x_train[numerical_cols] = sc.fit_transform(x_train[numerical_cols])

# Transform the test data using the fitted scaler
x_test[numerical_cols] = sc.transform(x_test[numerical_cols])
```
